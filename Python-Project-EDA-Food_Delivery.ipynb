```python

num_col = ['Price', 'Avg ratings', 'Total ratings', 'Delivery time']
num_col

```

```python
# Visualization(NUMERICAL COLUMN)
for col in num_col:
    plt.figure(figsize=(12, 6))
    plt.hist(df[col], bins=30, edgecolor='k',alpha=0.6)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

```

```python

cat_col = ['Area', 'City', 'Restaurant', 'Food type']
cat_col

```

```python
for col in cat_col:
    plt.figure(figsize=(12, 6))
    df[col].value_counts().head(10).plot(kind='bar', color='lightgreen', edgecolor='k')
    plt.title(f"Top Most Categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()
```

```python
#BIVARIATE ANALYSIS

#Scatter plot 
plt.figure(figsize=(10, 5))
plt.scatter(df['Price'], df['Avg ratings'], edgecolor='g')
plt.title("Price vs. Avg Ratings")
plt.xlabel("Price")
plt.ylabel("Avg Ratings")
plt.grid(True)
plt.show()
```

```python
#Insights>> with increase in prices ratings spread across the plot,
#increase or decrease in price does not produce higher ratings.
```

```python
# Food Type vs. Avg Ratings
t_food = df['Food type'].value_counts().head(10).index
t_food_rows = df[df['Food type'].isin(t_food)]


t_food_rows.boxplot(column='Avg ratings', by='Food type', figsize=(12, 6), grid=False)
plt.title("Food Type vs. Avg Ratings")
plt.xlabel("Food Type")
plt.ylabel("Avg Ratings")
plt.xticks(rotation=45)
plt.suptitle("") 
plt.show()
```

```python
#insights>>South Indian Food has the highest median ratings
#followed by the combination of North Indian ,Chinese and Desserts
```

```python

t_area =df.groupby('Area')['Avg ratings'].mean().sort_values(ascending=False).head(10)


plt.figure(figsize=(10, 6))
t_area.plot(kind='bar', color='skyblue', edgecolor='k')
plt.title("Top 10 Areas with Highest Average Ratings")
plt.xlabel("Area")
plt.ylabel("Avg Ratings")
plt.xticks(rotation=45)
plt.show()

```

```python
#Insights>>Girish Park has the highest average ratings
```

```python
t_rest = df.groupby('Restaurant')['Total ratings'].sum().sort_values(ascending=False).head(10)

# Plot for Top 10 Restaurants by Total Ratings
plt.figure(figsize=(10, 6))
t_rest.plot(kind='bar', color='lightgreen', edgecolor='k')
plt.title("Top 10 Restaurants with Highest Total Ratings")
plt.xlabel("Restaurant")
plt.ylabel("Total Ratings")
plt.xticks(rotation=45)
plt.show()
```

```python
#Insights>>SHAH Ghouse Cafe and Restaurant has the highest total ratings
```
