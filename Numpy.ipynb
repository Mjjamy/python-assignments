```python
#A2-Comparison of np.mean() and np.average() in NumPy:

#1. np.mean() calculates the arithmetic mean (average) of an array’s elements. It treats all elements equally, without weights.

#2. np.average() can calculate a weighted average, allowing you to assign different importance (weights) to each element.

#3. If no weights are given, np.average() works like np.mean().

#4. Use np.mean() when you want the simple average of all values.

#5. Use np.average() when you need to compute a weighted average.

#Example with np.mean():
import numpy as np
data = [1, 2, 3, 4, 5]
mean_value = np.mean(data)
print(mean_value)

```

```python
# Example with np.average() (weighted average):

import numpy as np
data = [1, 2, 3, 4, 5]
weights = [1, 2, 3, 4, 5]
average_value = np.average(data, weights=weights)
print(average_value)


#In summary, use np.mean() for a simple average and np.average() when weights are needed.
```

```python
#A3- the methods for reversing a NumPy array along different axes. Provide examples for 1D and 2D
#arrays
#Methods for Reversing a NumPy Array:
#For 1D array: Use [::-1] to reverse the array.

#For 2D array: Use [::-1, :] to reverse along rows (axis 0).

#For 2D array: Use [:, ::-1] to reverse along columns (axis 1).

#np.flip(array, axis): Flips along the specified axis.

#np.fliplr() and np.flipud(): Flip left-right and up-down for 2D arrays.

#Example of 1d array:
import numpy as np
arr_1d=np.array([1,2,3,4,5])
reversed_1d=arr_1d[::-1]
print(reversed_1d)
```

```python
#A7-Compare vstack() and hstack() functions in NumPy. Provide examples demonstrating their usage and  output

#1-np.vstack(): Stacks arrays vertically (row-wise), adding new rows to the existing array while keeping the number of columns the same.

#2-np.hstack(): Stacks arrays horizontally (column-wise), adding new columns to the existing array while keeping the number of rows the same.

#3-Both functions require that the dimensions of the arrays match appropriately for the stacking operation

#4-Example of vstack()
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6]])
result_vstack= np.vstack((arr1,arr2))
print(result_vstack)
```

```python
#5-Example of hstack()
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5],[6]])
result_hstack=np.hstack((arr1,arr2))
print(result_hstack)
```

```python
#A8-Explain the differences between fliplr() and flipud() methods in NumPy, including their effects on various  array dimensions.

#1-np.fliplr(): Flips an array from left to right (horizontally), reversing the order of the columns for 2D arrays.

#2np.flipud(): Flips an array from up to down (vertically), reversing the order of the rows for 2D arrays.

#3-Both methods do not modify the original array; they return a new array with the applied transformations.

#4-For 1D arrays, both functions will not have any effect since there are no rows or columns to flip

#5-example
arr=np.array([[1,2,3],[4,5,6]])
print(np.fliplr(arr))
```

```python
arr=np.array([[1,2,3],[4,5,6]])
print(np.flipud(arr))
```

```python
#A9- Discuss the functionality of the array_split() method in NumPy. How does it handle uneven splits

#1-The array_split() method in NumPy divides an array into multiple sub-arrays along a specified axis,
#allowing for more flexibility than split().

#2-It can handle uneven splits, meaning sub-arrays can have different sizes
# when the original array's length is not perfectly divisible by the number of splits.

#3-You can specify the axis along which to split, with the default being axis 0 (rows).

#4-It returns a list of the resulting sub-arrays, making it easy to access and manipulate each part.

#5-Example
arr=np.array([1,2,3,4,5,6,7])
split=np.array_split(arr,3)
print(split)
```

```python
#A10- Explain the concepts of vectorization and broadcasting in NumPy. How do they contribute to efficient array  operations

#Vectorization
#1-Definition: Vectorization allows operations on entire NumPy arrays without explicit loops, enabling element-wise computations efficiently.

#2-Efficiency: It reduces execution time by utilizing optimized, low-level code, making operations faster compared to interpreted Python loops.

#3-example of vectorization
a=np.array([1,2,3])
b=np.array([4,5,6])
result=a+b
print(result)
```

```python
#Broadcasting
#1-Definition: Broadcasting enables arithmetic operations on arrays of different shapes by automatically
#expanding the smaller array to match the larger array's shape.

#2-example of broadcasting
arr=np.array([1,2,3])
result=arr+5
print(result)
```
