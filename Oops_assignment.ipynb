```python
#A4-Python implement method overloading>>two or more methods in a class with the same name and different arguements(or parameters)
#is used to add more to the behaviour of methods and there is no need of more than one class for method overloading
#example
class A:
  def add(self,n1=None,n2=None,n3=None):

    if n1 is not None and n2 is not None and n3 is not None:
      result=n1+n2+n3
    elif n1 is not None and n2 is not None:
      result=n1+n2
    else:
      result=n1

    return result
a1=A()
print(a1.add(1,2))    #add method can take different forms within the same class
```

```python
#A6-Inheritance>>refers to the  process of child class
#syntax
#class baseclass:
  #body of base class
#class derivedclass(baseclass):
  #body of derived class

#types
#1-single inheritance>>when the derived class has only one parent class
#example
class father:
  def father_prop(self):
    print("this is father's property")
class son(father):
  def job(self):
    print("Son has property from job")
child_obj=son()    #a child can have his property as well as his fathers property
child_obj.job()
child_obj.father_prop()
```

```python
#3-multiple inheritance>>one child class may inherit the properties of multiple parent class
#example
class parentclass1:
  def method_1(self):
    print("method 1 of parent class1")
class parentclass2:
  def method_2(self):
    print("method 2 of parent class2")
class childclass(parentclass1,parentclass2):
  def method(self):
    print("method of child class")
c1=childclass()
c1.method()

```

```python
#A7-Method Resolution Order(MRO) in Python
#example>>diamond prob.>>it occurs when a class inherits from 2 or more than 2 classes>>so it leads to ambiguity in execution of methods
#to remove diamond problem>>python uses method resolution order(MRO) called c3 linearization

class parentclass1:
  def method_par(self):
    print("method 1 of parent class1")
class parentclass2:
  def method_par(self):
    print("method 2 of parent class2")
class childclass(parentclass1,parentclass2):   #meaning the class that is inherited first in the derived class,that method will be called
  def method(self):                            #in this case method of parentclass1
    print("method of child class")
a=childclass()
a.method_par()

```

```python
#code
class bank_account:
  def __init__(self,balance,accountno):
    self.__balance=balance
    self.__accountno=accountno

  def deposit(self,amount):
    self.__balance=self.__balance+amount

  def withdraw(self,amount):
    if self.__balance>=amount:
      self.__balance=self.__balance-amount
      return True
    else:
      return False

  def bal_inquiry(self):
    return self.__balance
```

```python
#A12- a decorator that measures and prints the execution time of a function.
#code
import time

def timer(func):
    def wrapper(*args, **kwargs):
        nonlocal total
        start = time.time()
        result = func(*args, **kwargs)
        duration = time.time() - start
        total += duration
        print(f"Execution time: {duration}   Total: {total}")
        return result

    total = 0
    return wrapper
```
