```python
#A1-Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each
#method and the final output after each method
#Here are three different methods to create identical 2D arrays in NumPy, along with simple short code examples for each method:

### Method 1: Using np.array()
#You can create a 2D array by directly passing a list of lists to the `np.array()` function.


import numpy as np

# Method 1: Using np.array()
array_1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Array created using np.array():\n", array_1)



```

```python
### Method 2: Using np.zeros()
#You can create a 2D array of zeros and then fill it with your desired values.


# Method 2: Using np.zeros()
array_2 = np.zeros((2, 3), dtype=int)  # Create a 2D array of zeros
array_2[0] = [1, 2, 3]  # Fill first row
array_2[1] = [4, 5, 6]  # Fill second row
print("Array created using np.zeros():\n", array_2)



```

```python
### Method 3: Using np.ones()
#You can create a 2D array of ones and then modify it to contain the desired values.


# Method 3: Using np.ones()
array_3 = np.ones((2, 3), dtype=int)  # Create a 2D array of ones
array_3[0] = [1, 2, 3]  # Fill first row
array_3[1] = [4, 5, 6]  # Fill second row
print("Array created using np.ones():\n", array_3)


### Summary
#Method  uses np.array() for direct creation.
#Method  utilizes `np.zeros()` to create an array of zeros and fills it with desired values.
#Method  leverages `np.ones()` to create an array of ones and modifies it to meet requirements.
```

```python
#A3-Explain the following terms
#The difference in np.array, np.asarray and np.asanyarray
#The difference between Deep copy and shallow copy
#1. Differences among np.array(), np.asarray(), and np.asanyarray():

#np.array(): Creates a new array from input data, always returning a fresh copy.

#np.asarray(): Converts input to an array but returns the same array if the input is already an array (no new copy).

#np.asanyarray(): Similar to np.asarray(), but keeps subclass arrays (like masked arrays) unchanged.

#2. Difference between Deep Copy and Shallow Copy:

#Shallow Copy: Creates a new object but keeps references to the original data. Changes to the original affect the copy.

#Deep Copy: Creates a completely new object with its own data. Changes to the original do not affect the copy.
```

```python
#A14- Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
#Perform the following tasks using Matplotlib
#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temerature' and  right y-axis for 'Humidity').

#b) Label the x-axis as 'Date'.

#c) Set the title of the plot as 'Temperature and Humidity Over Time'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
date_range = pd.date_range(start='2024-01-01', periods=30, freq='D')

# Generate random data for Temperature and Humidity
temperature = np.random.uniform(20, 35, size=30)  # Random temperatures between 20 and 35 degrees
humidity = np.random.uniform(40, 70, size=30)     # Random humidity between 40% and 70%

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Plotting Temperature and Humidity with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```
