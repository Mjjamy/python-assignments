```python
#1-What is a parameter?
#Ans-
#-Parameters are values the model learns from the data during training (e.g., weights in a model).

#-Parameters help transform features, like scaling or encoding, to make them useful for the model.

#-In Optimization-Parameters are adjusted using algorithms like gradient descent to minimize the loss function and improve model predictions.
```

```python
#2-What is correlation?
#What does negative correlation mean?

#Ans-Correlation: It measures how strongly two variables are related to each other.

#Positive Correlation: When one variable increases, the other also increases (e.g., height and weight).

#Negative Correlation: When one variable increases, the other decreases (e.g., temperature and sales of winter jackets).
```

```python
#3-Define Machine Learning. What are the main components in Machine Learning
#Machine Learning is a way for computers to learn patterns from data and make predictions or decisions without being explicitly programmed.

#3 Main Components:

#Data: The information used for learning.
#Model: The system or algorithm that learns patterns.
#Training: The process of teaching the model using data.


#Goal: To make accurate predictions or automate tasks based on the learned patterns.
```

```python
#4-How does loss value help in determining whether the model is good or not

#Loss Value: It shows how far the model's predictions are from the actual values; a lower loss means better predictions.

#If the loss value is high, the model is not performing well and may need adjustments or improvements.
```

```python
#5-What are continuous and categorical variables
#Continuous Variables: These are numbers that can take any value within a range (e.g., height, weight, or temperature).

#Categorical Variables: These are labels or categories that represent groups (e.g., colors like red/blue, or types like car/bike)
```

```python
#6-How do we handle categorical variables in Machine Learning? What are the common techniques
#Encoding: Convert categories into numbers using techniques like One-Hot Encoding (creates a separate column for each category)
#or Label Encoding (assigns a unique number to each category).

#Special Methods: For advanced models, use techniques like Target Encoding
#(replace categories with the mean of the target variable)
```

```python
#7-What do you mean by training and testing a dataset
#Training Dataset: It’s the part of the data used to teach the model by showing it patterns and relationships.

#Testing Dataset: It’s a separate part of the data used to check how well the model learned and performs on unseen data
```
