```python
#1-Can we use Bagging for regression problems

#Yes, Bagging can be used for regression problems.

# It combines predictions from multiple models (like decision trees) to reduce variance and improve accuracy.
```

```python
#2- What is the difference between multiple model training and single model training

#Single model training uses one algorithm to make predictions, relying on its performance alone.

# Multiple model training (ensemble) combines several models to improve accuracy and reduce errors
```

```python
#3- Explain the concept of feature randomness in Random Forest

#Feature randomness in Random Forest means each tree selects a random subset of features at each split.

#  This adds diversity among trees, reducing overfitting and improving generalization.
```

```python
#4- What is OOB (Out-of-Bag) Score

#OOB (Out-of-Bag) score is an internal validation method in Random Forest using data not included in each tree's bootstrap sample.

#  It estimates model accuracy without needing a separate test set.
```

```python
#5-How can you measure the importance of features in a Random Forest model

#Feature importance in Random Forest is measured by how much each feature decreases impurity (like Gini or MSE) across all trees.

# Higher decrease means the feature is more important for making accurate predictions
```

```python
#6- Explain the working principle of a Bagging Classifier

#A Bagging Classifier builds multiple models on different bootstrap samples of the data and combines their predictions (majority vote for classification).

#  This reduces variance and improves model stability and accuracy
```

```python
#7- How do you evaluate a Bagging Classifier’s performance

#You evaluate a Bagging Classifier’s performance using metrics like accuracy, precision, recall, and F1-score on a test set.

#  Cross-validation or Out-of-Bag (OOB) score can also assess its generalization ability.
```

```python
#8-How does a Bagging Regressor work

#A Bagging Regressor trains multiple regressors on different bootstrap samples and averages their predictions.

#  This reduces variance and improves prediction stability and accuracy.
```

```python
#9-What is the main advantage of ensemble techniques

#The main advantage of ensemble techniques is they combine multiple models to improve overall performance and reduce overfitting.

#  This leads to more accurate and robust predictions than individual models.
```

```python
#10-What is the main challenge of ensemble methods

#The main challenge of ensemble methods is increased computational complexity and training time.

#  They can also be harder to interpret compared to single models.
```

```python
#11- Explain the key idea behind ensemble techniques

#The key idea behind ensemble techniques is to combine multiple models to leverage their strengths and reduce individual weaknesses.

#  This results in improved accuracy, robustness, and generalization.
```

```python
#12- What is a Random Forest Classifier

#A Random Forest Classifier is an ensemble learning method that creates multiple decision trees using random subsets of data and features.

#  It combines their predictions through majority voting to improve classification accuracy.
```

```python
#13-What are the main types of ensemble techniques

#The main types of ensemble techniques are bagging, boosting, and stacking. 

# Bagging reduces variance, boosting improves accuracy by focusing on errors, and stacking combines different models for better predictions.
```

```python
#14-What is ensemble learning in machine learning

#Ensemble learning is a machine learning technique that combines multiple models to improve prediction accuracy and robustness.

#  It leverages the diversity of models to reduce errors and increase overall performance.
```

```python
#15- When should we avoid using ensemble methods

#Ensemble methods should be avoided when model interpretability is crucial, as they can be complex and harder to explain.

#  They are also not ideal for small datasets, where simpler models may perform equally well with less computational cost.
```

```python
#16- How does Bagging help in reducing overfitting

#Bagging reduces overfitting by averaging predictions from multiple models trained on different data subsets, which smooths out fluctuations.

# This helps decrease variance without sacrificing bias, making the model more generalizable.
```

```python
#17-Why is Random Forest better than a single Decision Tree

#Random Forest is better than a single Decision Tree because it aggregates multiple trees, reducing variance and overfitting.

#  This results in more accurate and robust predictions compared to a single, prone-to-overfitting tree
```

```python
#18- What is the role of bootstrap sampling in Bagging

#Bootstrap sampling in Bagging involves creating multiple random subsets of the data with replacement,

# allowing each model to train on different data points. 

# This increases diversity among models, helping to reduce variance and improve accuracy.
```

```python
#19- What are some real-world applications of ensemble techniques

#Ensemble techniques are used in real-world applications like spam email detection, fraud detection, and predictive maintenance.

#  They also power recommendation systems, medical diagnoses, and stock market predictions.
```

```python
#20-What is the difference between Bagging and Boosting

#Bagging trains multiple models independently on random subsets of data and combines their predictions, reducing variance.

# Boosting trains models sequentially, focusing on correcting errors from previous models, improving accuracy by reducing bias.
```

```python
#PRACTICAL

#21 Train a Bagging Classifier using Decision Trees on a sample dataset and print model accuracy

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


data = load_iris()
X = data.data
y = data.target


```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


bagging_model = BaggingClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=50, random_state=1)


bagging_model.fit(X_train, y_train)


y_pred = bagging_model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
#22- Train a Bagging Regressor using Decision Trees and evaluate using Mean Squared Error (MSE)

from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


X, y = make_regression(n_samples=100, n_features=5, noise=0.1, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#24-Train a Random Forest Regressor and compare its performance with a single Decision Tree

from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


X, y = make_regression(n_samples=100, n_features=5, noise=0.1, random_state=1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

dt_regressor = DecisionTreeRegressor(random_state=1)
dt_regressor.fit(X_train, y_train)



```

```python
#25-Compute the Out-of-Bag (OOB) Score for a Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split


data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)


rf_classifier = RandomForestClassifier(n_estimators=100, oob_score=True, random_state=1)
```

```python
#27- Train a Random Forest Classifier with different numbers of trees and compare accuracy

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


data = load_iris()
X = data.data
y = data.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)


n_trees_list = [10, 50, 100, 200, 500]


accuracy_dict = {}


```

```python
for n_trees in n_trees_list:
    rf_classifier = RandomForestClassifier(n_estimators=n_trees, random_state=1)
    rf_classifier.fit(X_train, y_train)
    
    
    y_pred = rf_classifier.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_dict[n_trees] = accuracy


for n_trees, accuracy in accuracy_dict.items():
    print(f"Number of Trees: {n_trees} -> Accuracy: {accuracy * 100:.2f}%")
```

```python
#28- Train a Bagging Classifier using Logistic Regression as a base estimator and print AUC score

from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize


data = load_iris()
X = data.data
y = data.target


y_bin = label_binarize(y, classes=[0, 1, 2])


X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=1)


log_reg_estimator = LogisticRegression(solver='liblinear', random_state=1)


```

```python

bagging_lr = BaggingClassifier(base_estimator=log_reg_estimator, n_estimators=50, random_state=1)


bagging_lr.fit(X_train, y_train)

y_pred_prob = bagging_lr.predict_proba(X_test)


auc_scores = [roc_auc_score(y_test[:, i], y_pred_prob[i][:, 1]) for i in range(y_test.shape[1])]


for i, auc in enumerate(auc_scores):
    print(f"AUC for class {i}: {auc:.4f}")
```

```python
#30-Train an ensemble model using both Bagging and Random Forest and compare accuracy

from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


data = load_iris()
X = data.data
y = data.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

base_classifier = DecisionTreeClassifier(random_state=1)


```

```python
bagging_classifier = BaggingClassifier(base_estimator=base_classifier, n_estimators=50, random_state=1)
bagging_classifier.fit(X_train, y_train)


random_forest_classifier = RandomForestClassifier(n_estimators=50, random_state=1)
random_forest_classifier.fit(X_train, y_train)


y_pred_bagging = bagging_classifier.predict(X_test)
y_pred_rf = random_forest_classifier.predict(X_test)


accuracy_bagging = accuracy_score(y_test, y_pred_bagging)
accuracy_rf = accuracy_score(y_test, y_pred_rf)


print(f"Accuracy of Bagging Classifier: {accuracy_bagging * 100:.2f}%")
print(f"Accuracy of Random Forest Classifier: {accuracy_rf * 100:.2f}%")
```

```python
#32- Train a Bagging Regressor with different numbers of base estimators and compare performance

from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

data = fetch_california_housing()
X = data.data
y = data.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

base_estimator = DecisionTreeRegressor(random_state=1)


n_estimators_list = [10, 50, 100, 200, 500]


mse_dict = {}



```

```python
for n_estimators in n_estimators_list:
    bagging_regressor = BaggingRegressor(base_estimator=base_estimator, n_estimators=n_estimators, random_state=1)
    bagging_regressor.fit(X_train, y_train)
    
   
    y_pred = bagging_regressor.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    mse_dict[n_estimators] = mse


for n_estimators, mse in mse_dict.items():
    print(f"Number of Estimators: {n_estimators} -> MSE: {mse:.4f}")


plt.figure(figsize=(10, 6))
plt.plot(n_estimators_list, list(mse_dict.values()), marker='o')
plt.xlabel("Number of Base Estimators")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Bagging Regressor - Performance with Different Numbers of Estimators")
plt.grid(True)
plt.show()
```

```python
#33-Train a Random Forest Classifier and analyze misclassified samples

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd


data = load_iris()
X = data.data
y = data.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)


rf_classifier = RandomForestClassifier(n_estimators=100, random_state=1)


rf_classifier.fit(X_train, y_train)



```

```python
y_pred = rf_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Random Forest Classifier: {accuracy * 100:.2f}%")


misclassified_samples = X_test[y_pred != y_test]
true_labels = y_test[y_pred != y_test]
predicted_labels = y_pred[y_pred != y_test]


misclassified_df = pd.DataFrame(misclassified_samples, columns=data.feature_names)
misclassified_df['True Label'] = true_labels
misclassified_df['Predicted Label'] = predicted_labels


print("\nMisclassified Samples:")
print(misclassified_df)
```

```python
#34- Train a Bagging Classifier and compare its performance with a single Decision Tree Classifier

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


data = load_iris()
X = data.data
y = data.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

dt_classifier = DecisionTreeClassifier(random_state=1)


dt_classifier.fit(X_train, y_train)


y_pred_dt = dt_classifier.predict(X_test)


accuracy_dt = accuracy_score(y_test, y_pred_dt)

```

```python

bagging_classifier = BaggingClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=50, random_state=1)


bagging_classifier.fit(X_train, y_train)


y_pred_bagging = bagging_classifier.predict(X_test)


accuracy_bagging = accuracy_score(y_test, y_pred_bagging)


print(f"Accuracy of Decision Tree Classifier: {accuracy_dt * 100:.2f}%")
print(f"Accuracy of Bagging Classifier: {accuracy_bagging * 100:.2f}%")
```
