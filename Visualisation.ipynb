```python
#A5-Show a pie chart to represent the percentage distribution of different sections in the array sections.
#sections = ['Section A', 'Section B', 'Section C', 'Section D']
#sizes = [25, 30, 15, 30]

import matplotlib.pyplot as plt

# Given sections and their sizes
sections = ['Section A', 'Section B', 'Section C', 'Section D']
sizes = [25, 30, 15, 30]

# Create a pie chart
plt.pie(sizes, labels=sections, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title('Percentage Distribution of Sections')

plt.show()
```

```python
#SEABORN ASSIGNMENT
#A1- Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
#dataset.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset
np.random.seed(42)  # For reproducibility
n = 100
# Create two variables, x and y, with a clear relationship
x = np.linspace(1, 100, n)  # X values from 1 to 100
y = np.sqrt(x) + np.random.normal(0, 1, n)  # Y values based on the square root of X with some noise

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Create a scatter plot using Seaborn
sns.scatterplot(data=data, x='X', y='Y', color='blue')

# Add labels and a title
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Scatter Plot of Y vs. X')

# Show the plot
plt.show()
```

```python
#A2- Generate a dataset of random numbers. Visualize the distribution of a numerical variable
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a dataset of 1000 random numbers from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, StdDev=10

# Create a DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Visualize the distribution using Seaborn
sns.histplot(df['Value'], bins=30, kde=True, color='skyblue', edgecolor='black')

# Add labels and a title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Random Numbers')

# Show the plot
plt.grid()
plt.show()
```

```python
#A3. Create a dataset representing categories and their corresponding values. Compare different categories
#based on numerical values.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset representing categories and their corresponding values
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [25, 40, 30, 35, 20]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Visualize the comparison of different categories using a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Values', data=df, palette='viridis')

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Comparison of Categories Based on Values')

# Show the plot
plt.grid(axis='y')
plt.show()
```

```python
#Plotly Assignment
#A1-Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space. use plotly!
#np.random.seed(30)
#data = {
    #'X': np.random.uniform(-10, 10, 300),
    #'Y': np.random.uniform(-10, 10, 300),
    #'Z': np.random.uniform(-10, 10, 300)
#}
#df = pd.DataFrame(data)
import numpy as np
import pandas as pd
import plotly.express as px

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot using Plotly
fig = px.scatter_3d(df, x='X', y='Y', z='Z', title='3D Scatter Plot of Data Points',
                     color=df.index,  # Color by index for diversity
                     labels={'X': 'X Axis', 'Y': 'Y Axis', 'Z': 'Z Axis'})

# Show the plot
fig.show()
```
